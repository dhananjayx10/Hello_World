# Program for checking Applicant is shortlist or not ?

```python
def python_out_of(py):
    if 3 == py:
        return(10)
    elif 2 == py:
        return(7)
    elif 1 == py:
        return(3)
    else:
        return(py)

def r_out_of(r):
    if 3 == r:
        return(10)
    elif 2 == r:
        return(7)
    elif 1 == r:
        return(3)
    else:
        return(r)
    
    
def ds_out_of(ds):
    if 3 == ds:
        return(10)
    elif 2 == ds:
        return(7)
    elif 1 == ds:
        return(3)
    else:
        return(ds) 
    
def check(string, sub_str):
    if (string.find(sub_str) == -1): 
        return 0
    else: 
        return (5) 

def check_(string, sub_str):
    if (string.find(sub_str) == -1): 
        return 0
    else: 
        return (3) 
    
    
def year(yy):
    if yy == 2020:
        return (5)
    elif yy == 2019:
        return (3)
    else:
        return 0
    
for x in range(0,len(data.index)):
    print("Application ID No.",data['Application_ID'][x])
    
    d = (data["Degree"][x])
    be = check(d, "B.E")
    bt = check(d, "B.Tech")
    mt = check_(d, "M.Tech")
    ms = check_(d, "M.Sc")
    degree = be + bt + mt + ms
    
    yy = (data["Current Year Of Graduation"][x])
    cyof = year(yy)
    
    
    py = (data["Python (out of 3)"][x])
    python = python_out_of(py) 
    
    
    r = (data["R Programming (out of 3)"][x])
    r_programming = r_out_of(r)
    
    
    ds = (data["Data Science (out of 3)"][x])
    data_science = ds_out_of(ds)

    
    os = (data["Other skills"][x])
    x = [word for word in os.split(", ")]
    x1 = x.count("Machine Learning")
    x2 = x.count("Deep Learning")
    x3 = x.count("NLP")
    x4 = x.count("Statistical")
    x5 = x.count("AWS")
    x6 = x.count("SQL")
    x7 = x.count("Excel")
    x8 = x.count("NoSQL")
    skills = x1 + x2 + x3 + x4 + x5 + x6 + x7 + x8
    other_skills = (skills*3)
    total = degree + cyof + python + r_programming + data_science + other_skills
    if (total>=30):
        print("Congratulation !\nYour profile has been shortlisted for Data Scientist.")
    else:
        print("Sorry !\nYour profile did not qualify for further discussion.")
    
    print('\n')
    
```
